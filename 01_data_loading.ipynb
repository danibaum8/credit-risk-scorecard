```python
import pandas as pd

col_names = [
    'checking_account', 'duration', 'credit_history', 'purpose', 'credit_amount',
    'savings_account', 'employment', 'installment_rate', 'personal_status',
    'other_debtors', 'residence_since', 'property', 'age', 'other_installments',
    'housing', 'existing_credits', 'job', 'dependents', 'telephone', 'foreign_worker',
    'target'
]

df = pd.read_csv('data/german.data.txt', sep=' ', names=col_names)

print(df.shape)
df.head()
```

```python
print(df.dtypes)
print('\nMissing values:')
print(df.isnull().sum())
print('\nTarget distribution:')
print(df['target'].value_counts())
```

```python
import sqlite3
from sqlalchemy import create_engine

# Create SQLite database in your project folder
engine = create_engine('sqlite:///data/creditrisk.db')

# Load the dataframe into the database
df.to_sql('german_credit', engine, if_exists='replace', index=False)

print("Data loaded into SQLite successfully!")
print(f"Table: german_credit | Rows: {len(df)}")
```
